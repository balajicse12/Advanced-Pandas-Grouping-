```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

```python
print("\n" + "="*50)
print("Task 1: Basic Grouping and Single Aggregations")

# 1. Total sales per region
reg_sales = df.groupby('region')['sales_amount'].sum().reset_index()
reg_sales_sorted = reg_sales.sort_values(by='sales_amount', ascending=False)
print("\nTotal Sales per Region (Sorted):")
print(reg_sales_sorted)

# 2. Count transactions per category
cat_counts = df.groupby('category')['transaction_id'].count().reset_index()
print("\nTransactions per Category:")
print(cat_counts)

# 3. Average sales per salesperson
sales_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print("\nAverage Sales per Salesperson:")
print(sales_avg)
```

```python
print("\n" + "="*50)
print("Task 2: Multi-Column Grouping and Multiple Aggregations")

# 1. Group by Region AND Category
reg_cat_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
print("\nSales by Region and Category combination (Preview):")
print(reg_cat_sales.head())

# 2. Salesperson performance with multiple metrics
salesperson_perf = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()
salesperson_perf_sorted = salesperson_perf.sort_values(by='sum', ascending=False)
print("\nSalesperson Metrics (Sorted by Total Sales):")
print(salesperson_perf_sorted)

# 3. Find category with maximum revenue
cat_totals = df.groupby('category')['sales_amount'].sum()
top_category = cat_totals.idxmax()
print(f"\nCategory with Maximum Total Revenue: {top_category}")
```

```python
print("\n" + "="*50)
print("Task 3: Custom Aggregation and Complete Sales Report")

# 1. Define custom range function
def sales_range(x):
    return x.max() - x.min()

# 2. Apply custom and standard functions
reg_custom_analysis = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print("\nRegional Sales Analysis with Custom Range:")
print(reg_custom_analysis)

# 3. Final Summary Report using Dictionary Syntax
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("\nFinal Regional Summary Report:")
print(final_report)
```
